## **Visual Exploratory Data Analysis**

```python
plt.figure(figsize=(10,6))

# Add title
plt.title("average score for racing games")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=data.key,y=data.loudness)
# Add label for vertical axis
# Your code here
plt.title("Loudness value by the spotify recommendation")
```

Here the relation is in between the key and loudness of the given spotify data

```python
# Set the width and height of the figure
plt.figure(figsize=(14,35))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=data, annot=True)

# Add label for horizontal axis
plt.xlabel("Spotify Recommendation Dataset")
```

The above diagram is showing the average plot drawn by the loudness value of the spotify Recommendation
